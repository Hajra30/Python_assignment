## Cleaning and Filtering the data

```python
# Find missing values
ks.isnull()
```

```python
ks.isnull().sum()
```

```python
# dropping one with more missing values
ks.drop(['deck'],axis=1)
```

```python
ks_clean=ks.drop(['age','deck'],axis=1)
ks_clean
```

```python
ks_clean.shape
```

```python
ks_clean=ks_clean.dropna()
ks_clean
```

```python
ks_clean.shape
```

```python
ks_clean.isnull().sum()
```

```python
ks_clean['sex'].value_counts()
```

```python
ks_clean['fare'].value_counts()
```

```python
ks_clean.describe()
```

```python
sns.boxplot(x='sex',y='survived',data=ks_clean)
```

```python
sns.boxplot(x='who',y='pclass',data=ks_clean)
```

```python
sns.boxplot(x='class',y='fare',data=ks_clean)
```

```python
# elimnation of outliers
ks_clean=ks_clean[ks_clean['fare']<300]
ks_clean
```

```python
sns.boxplot(x='class',y='fare',data=ks_clean)
```
